## Linear Regression using Gradient Descent

```python
def mse(b,b_pred):
    return np.sum((b-b_pred)**2)/(len(b))
```

```python
mse(b,b_pred)
```

```python
def gradient_descent(slope,intercept):
    alpha = 0.01
    n = 1000
    cost = 0
    for i in range(n):
        b_pred = slope*a + intercept
        error = b - b_pred
        cost = (1/len(a)) * sum(error**2)
        a_diff = -(1/len(a)) * sum(error)
        b_diff = -(1/len(a)) * sum(error * a)
        intercept = intercept - alpha * a_diff
        slope = slope - alpha * b_diff

    return slope,intercept,cost
```

```python
m,c,cost=gradient_descent(m,c)
print(m)
print(c)
```

## Calculating Slope and Intercept using Formula

```python
mean_a = np.mean(a)
mean_b = np.mean(b)
```

```python
num = 0
den = 0
for i in range(len(a)):
  num += (a[i] - mean_a) * (b[i] - mean_b)
  den += (b[i] - mean_b) ** 2
m = num/ den
c = mean_b - (m * mean_a)
print(m)
print(c)
```

```python
max_a = np.max(a) 
min_a = np.min(b) 

A = np.linspace (min_a, max_a, 100)
B = c + m * A

plt.scatter(a, b)
plt.scatter(A, B)
plt.show()
```
